### 1.Positional Encoding（位置嵌入，位置编码）:

由于transformer没有循环神经网络的迭代操作，所以必须提供每个字的位置信息，才能识别语言中的顺序关系。  
**维度：**  
【max sequence len, embedding dimension】  
**论文中使用sine和cosine函数的线性变换来提供给模型位置信息：**  
https://www.zhihu.com/question/347678607/answer/864217252  
*大神的知乎解释  
*BERT等后期换用了Positional Embedding

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
def get_positional_encoding(max_seq_len, embed_dim):
    positional_embedding = np.array([
        [pos/np.power(10000, 2*i/embed_dim) for i in range(embed_dim)] if pos!=0 
        else np.zeros(embed_dim) for pos in range(max_seq_len)])
    positional_embedding[1:, 0::2] = np.sin(positional_embedding[1:, 0::2])
    positional_embedding[1:, 1::2] = np.cos(positional_embedding[1:, 1::2])
    
    return positional_embedding
```

```python
positional_encoding = get_positional_encoding(max_seq_len=100, embed_dim=16)
plt.figure(figsize=(10,10))
sns.heatmap(positional_encoding)
plt.title("Sinusoidal Function")
plt.xlabel("Hidden dimention")
plt.ylabel("Sequence dimention")
```
